# Loading the Data

```python
#Loading dummy data
filename='dummytext.txt'
file=open(filename,'rt')
text=file.read()
file.close()
```

## Cleaning The Data

```python
from nltk import sent_tokenize
Sentences=sent_tokenize(text)
print(Sentences[0])
```

```python
Sentences[0]=Sentences[0][3:]
print(Sentences[0])
```

```python
print(len(Sentences))
```

```python
from nltk import word_tokenize
Doc=[]
for sent in Sentences:
    Doc.append(word_tokenize(sent))  
```

```python
print(Doc[0])
```

```python
print(Doc[700])
```

```python
from nltk import SnowballStemmer
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
stemmer=SnowballStemmer('english')
DocMod=[]
for WordList in Doc:
    l=[]
    for word in WordList:
        if word.isalpha() and word not in stop_words:
            l.append(stemmer.stem(word))
    DocMod.append(l)        
print(DocMod[700])            
```

```python
DocList=[]
for WordList in DocMod:
    Doc=''
    for word in WordList:
        Doc+=word+' '
    Doc=Doc.strip()
    DocList.append(Doc)
print(len(DocList))
print(DocList[0])
```

## Encoding a Doc

### Using NLTK

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
# tokenize and build vocab
vectorizer.fit(DocList)
# summarize
#print(vectorizer.vocabulary_)
print(vectorizer.idf_)
print(len(vectorizer.vocabulary_))
print(len(vectorizer.idf_))
```

```python
vector = vectorizer.transform([DocList[0]])
# summarize encoded vector
print(vector.shape)
print(DocList[0])
print(vector)
```

### Using Keras

```python
from keras.preprocessing.text import Tokenizer
# define 5 documents
docs=DocList
# create the tokenizer
t = Tokenizer()
# fit the tokenizer on the documents
t.fit_on_texts(docs)
# summarize what was learned
print(t.word_counts)
print(t.document_count)
print(t.word_index)
print(t.word_docs)
# integer encode documents
encoded_docs = t.texts_to_matrix(docs, mode='count')
print(encoded_docs)
```
